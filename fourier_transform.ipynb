```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
```

```python
def target_fn(x):
  return np.sin(2*pi*3*x) + np.sin(2*pi*5*x) + np.sin(2*pi*7*x) + np.sin(2*pi*9*x)
```

```python
X=np.linspace(0,11,10000)
#f=1
#Y=[1]*1000+[0]*1000+[1]*1000+[0]*1000+[1]*1000+[0]*1000+[1]*1000+[0]*1000+[1]*1000+[0]*1000
Y=target_fn(X)
plt.plot(X,Y)
```

```python
y=[]
for i in X:
  y.append(np.dot(Y,np.cos(2*pi*(i)*X)))

```

```python
plt.plot(X,y)
```

```python
y_o=[]
for i in X:
  y_o.append(np.dot(y,np.sin(2*pi*(i)*X)))

plt.plot(X,y_o)
```

```python
d={}
print(max(y))
for _ in range(10000):
  imp=max(y)
  imp_index=y.index(imp)
  freq=X[imp_index]
  d[freq]=imp
  y.remove(imp)

```
